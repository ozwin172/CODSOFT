Now, let's create box plots for each numerical column to visualize their distribution and identify any outliers. Outliers will appear as individual points beyond the 'whiskers' of the box plot.

We will use the **Interquartile Range (IQR) method** to identify and cap outliers in the numerical columns: 'Time (min)', 'Pages Viewed', and 'Amount Spent ($)'.

**Steps:**
1.  Calculate the First Quartile (Q1) and Third Quartile (Q3).
2.  Calculate the Interquartile Range (IQR = Q3 - Q1).
3.  Define the upper bound as Q3 + 1.5 * IQR and the lower bound as Q1 - 1.5 * IQR.
4.  Any value above the upper bound will be replaced with the upper bound, and any value below the lower bound will be replaced with the lower bound.

```python
import numpy as np

# Columns to handle outliers for (excluding 'Customer' as it's an ID)
columns_to_cap = ['Time (min)', 'Pages Viewed', 'Amount Spent ($)']

# Create a copy of the DataFrame to store the processed data
data_capped = clean_data.copy()

print("Original statistics before capping outliers:")
display(data_capped[columns_to_cap].describe())

for col in columns_to_cap:
    Q1 = data_capped[col].quantile(0.25)
    Q3 = data_capped[col].quantile(0.75)
    IQR = Q3 - Q1

    # Define outlier bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Cap the outliers
    data_capped[col] = np.where(data_capped[col] < lower_bound, lower_bound, data_capped[col])
    data_capped[col] = np.where(data_capped[col] > upper_bound, upper_bound, data_capped[col])

print(f"\nStatistics after capping outliers for columns: {columns_to_cap}")
display(data_capped[columns_to_cap].describe())

print("\nFirst 5 rows of the DataFrame after outlier capping:")
display(data_capped.head())
```

Now that the outliers have been capped, let's visualize the distributions again using box plots to see the effect of this operation. We should observe that the extreme points are no longer present, as they have been brought within the defined bounds.

```python
# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# Determine the number of rows and columns for the subplot grid
num_numerical_cols = len(columns_to_cap)
n_cols = 2  # You can adjust this for more columns per row
n_rows = (num_numerical_cols + n_cols - 1) // n_cols # Calculate rows needed

plt.figure(figsize=(12, n_rows * 5))

for i, col in enumerate(columns_to_cap):
    plt.subplot(n_rows, n_cols, i + 1)
    sns.boxplot(y=data_capped[col])
    plt.title(f'Box Plot of {col} (Outliers Capped)')
    plt.ylabel(col)

plt.tight_layout()
plt.show()

# Update clean_data to reflect the capped data
clean_data = data_capped.copy()
```

The box plots now show the distributions after capping the outliers. ## Rebuilding Linear Regression Models

We are re-building two simple linear regression models to predict 'Amount Spent ($)':
1.  Using 'Time (min)' as the independent variable.
2.  Using 'Pages Viewed' as the independent variable.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Define target variable
y = clean_data['Amount Spent ($)']

print("Target variable (y) defined: 'Amount Spent ($)'")
```

### Model 1: Predicting Amount Spent using Time (min)

### Model 2: Predicting Amount Spent using Pages Viewed

Training and evaluating the linear regression model with 'Pages Viewed'.

```python
# Feature variable for Model 2
X2 = clean_data[['Pages Viewed']]

# Split data into training and testing sets
# Using the same random_state for reproducibility
X2_train, X2_test, y_train, y_test = train_test_split(X2, y, test_size=0.2, random_state=42)

# Initialize and train the model
model_pages = LinearRegression()
model_pages.fit(X2_train, y_train)

# Make predictions on the test set
y_pred_pages = model_pages.predict(X2_test)

# Evaluate the model
mae_pages = mean_absolute_error(y_test, y_pred_pages)
mse_pages = mean_squared_error(y_test, y_pred_pages)
rmse_pages = np.sqrt(mse_pages)
r2_pages = r2_score(y_test, y_pred_pages)

print("--- Model 2: Predicting Amount Spent using Pages Viewed ---")
print(f"Mean Absolute Error (MAE): {mae_pages:.2f}")
print(f"Mean Squared Error (MSE): {mse_pages:.2f}")
print(f"Root Mean Squared Error (RMSE): {rmse_pages:.2f}")
print(f"R-squared (R2): {r2_pages:.2f}")

print(f"Model Coefficients: {model_pages.coef_[0]:.2f}")
print(f"Model Intercept: {model_pages.intercept_:.2f}")
```

## ANOVA Test: 'Amount Spent ($)' by Day of the Week

We will use a one-way ANOVA test to check if there is a statistically significant difference in the mean 'Amount Spent ($)' among the different days of the week.

```python
from scipy import stats

# Get unique days of the week
days = clean_data['Day'].unique()

# Create a list of 'Amount Spent ($)' for each day
amount_spent_by_day = [clean_data['Amount Spent ($)'][clean_data['Day'] == day] for day in days]

# Perform one-way ANOVA
f_statistic, p_value = stats.f_oneway(*amount_spent_by_day)

print(f"ANOVA F-statistic: {f_statistic:.2f}")
print(f"ANOVA p-value: {p_value:.3f}")

if p_value < 0.05:
    print("\nSince the p-value is less than 0.05, we reject the null hypothesis.")
    print("This suggests that there is a statistically significant difference in the mean 'Amount Spent ($)' across at least some of the days of the week.")
else:
    print("\nSince the p-value is greater than or equal to 0.05, we fail to reject the null hypothesis.")
    print("This suggests that there is no statistically significant difference in the mean 'Amount Spent ($)' across the different days of the week.")
```

### Interpretation

*   **F-statistic:** This value measures the ratio of variance between the groups (days) to the variance within the groups. A larger F-statistic suggests greater differences between group means.
*   **p-value:** This is the probability of observing an F-statistic as extreme as, or more extreme than, the one calculated, assuming the null hypothesis is true (i.e., there is no difference in means across days).

If the p-value is sufficiently small (typically less than 0.05), we conclude that there is a statistically significant difference in mean 'Amount Spent ($)' among the days of the week. However, ANOVA does not tell us *which* specific days differ; for that, post-hoc tests (like Tukey's HSD) would be needed.

## Post-hoc Test: Tukey's HSD for 'Amount Spent ($)' by Day of the Week

Even though the one-way ANOVA test did not find a statistically significant overall difference, a post-hoc test can provide more granular insights into pairwise comparisons. We will use Tukey's HSD to compare the mean 'Amount Spent ($)' for all possible pairs of days.

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import pandas as pd

# Perform Tukey's HSD post-hoc test
tukey_result = pairwise_tukeyhsd(endog=clean_data['Amount Spent ($)'],
                                 groups=clean_data['Day'],
                                 alpha=0.05)

print("Tukey's HSD Post-hoc Test Results:")
print(tukey_result)
```

### Interpretation of Tukey's HSD Results

*   **`meandiff`**: The difference between the means of the two groups being compared.
*   **`p-adj`**: The adjusted p-value for the comparison. If this value is less than the significance level (alpha, typically 0.05), then the difference between the means of that pair of groups is considered statistically significant.
*   **`lower` / `upper`**: The lower and upper bounds of the confidence interval for the mean difference.
*   **`reject`**: A boolean value indicating whether the null hypothesis for that specific pairwise comparison (that the means are equal) is rejected (True) or not (False).

### ANOVA Test for 'Amount Spent ($)' by Browser

Now, let's perform an ANOVA test for 'Amount Spent ($)' across different browsers to see if there's a statistically significant difference in mean spending.

```python
from scipy import stats

browsers = clean_data['Browser'].unique()
amount_spent_by_browser = [clean_data['Amount Spent ($)'][clean_data['Browser'] == browser] for browser in browsers]

f_statistic_amount, p_value_amount = stats.f_oneway(*amount_spent_by_browser)

print(f"ANOVA for Amount Spent ($) by Browser:")
print(f"F-statistic: {f_statistic_amount:.2f}")
print(f"p-value: {p_value_amount:.3f}")

if p_value_amount < 0.05:
    print("Conclusion: Statistically significant difference in mean Amount Spent ($) across browsers.")
else:
    print("Conclusion: No statistically significant difference in mean Amount Spent ($) across browsers.")
```

### ANOVA Test for 'Time (min)' by Browser

Next, we'll perform an ANOVA test for 'Time (min)' across different browsers.

```python
time_by_browser = [clean_data['Time (min)'][clean_data['Browser'] == browser] for browser in browsers]

f_statistic_time, p_value_time = stats.f_oneway(*time_by_browser)

print(f"\nANOVA for Time (min) by Browser:")
print(f"F-statistic: {f_statistic_time:.2f}")
print(f"p-value: {p_value_time:.3f}")

if p_value_time < 0.05:
    print("Conclusion: Statistically significant difference in mean Time (min) across browsers.")
else:
    print("Conclusion: No statistically significant difference in mean Time (min) across browsers.")
```

### ANOVA Test for 'Pages Viewed' by Browser

Finally, an ANOVA test for 'Pages Viewed' across different browsers.

```python
pages_by_browser = [clean_data['Pages Viewed'][clean_data['Browser'] == browser] for browser in browsers]

f_statistic_pages, p_value_pages = stats.f_oneway(*pages_by_browser)

print(f"\nANOVA for Pages Viewed by Browser:")
print(f"F-statistic: {f_statistic_pages:.2f}")
print(f"p-value: {p_value_pages:.3f}")

if p_value_pages < 0.05:
    print("Conclusion: Statistically significant difference in mean Pages Viewed across browsers.")
else:
    print("Conclusion: No statistically significant difference in mean Pages Viewed across browsers.")
```
